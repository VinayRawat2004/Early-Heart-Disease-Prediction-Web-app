```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv")
```

```python
print("Missing values in training data  before cleanup:\n", train.isnull().sum())
print("Missing values in testing data before cleanup:\n", test.isnull().sum())
```

```python
train.head()
```

```python
train.tail()
```

FILLING IN TRAIN DATA

```python
for col in train.columns:
    if train[col].dtype == "object":  # Categorical data
        train[col] = train[col].fillna(train[col].mode()[0])  # Fill with mode
    else:  # Numerical data
        train[col] = train[col].fillna(train[col].mean())  # Fill with mean
```

FOR TEST DATA

```python
for col in test.columns:
    if test[col].dtype == "object":  # Categorical data
        test[col] = test[col].fillna(test[col].mode()[0])  # Fill with mode
    else:  # Numerical data
        test[col] = test[col].fillna(test[col].mean())  # Fill with mean
```

```python
print("Missing values in training data  after cleanup:\n", train.isnull().sum())
print("Missing values in testing data after cleanup:\n", test.isnull().sum())
```

```python
categorical_columns = train.select_dtypes(include=['object']).columns
```

```python
for col in categorical_columns:
    print(f"Unique values in {col} (Train):", train[col].unique())
    print(f"Unique values in {col} (Test):", test[col].unique())
```

```python
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    label_encoders[col] = le
```

```python
train.head()
```

```python
import numpy as np

for col in categorical_columns:
    if col in label_encoders:
        le = label_encoders[col]
        test[col] = np.where(test[col].isin(le.classes_), le.transform(test[col]), -1)
```
